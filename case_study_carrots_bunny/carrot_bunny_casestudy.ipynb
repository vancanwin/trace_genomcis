```python
import random, os, math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# matrix dimensions
N = 4
M = 4

# generating garden of carrots

df = pd.DataFrame(np.random.randint(0,10,size=(N, M)))

#finding the center part of the matrix

# n_center = (N/2)
# m_center = (M/2)

n_center = math.ceil(N/2)
m_center = math.ceil(M/2)

# if the matrix centers are even then return 2 x 2 matrix and make sure the center_numbers don't have repeats

if n_center % 2 == 0 and m_center % 2 == 0:
    center_numbers = df.iloc[n_center-1:n_center+1, m_center-1:m_center+1]    
    random_center = np.array(random.sample(range(10), (center_numbers.shape[0]*center_numbers.shape[1])))
    df.iloc[n_center-1:n_center+1, m_center-1:m_center+1] = random_center.reshape(center_numbers.shape[0],center_numbers.shape[1])
    center_numbers = df.iloc[n_center-1:n_center+1, m_center-1:m_center+1]
    starter_carrot = np.max(center_numbers.values)

    coordinates={center_numbers.values[0][0]:[n_center-1,m_center-1],
                 center_numbers.values[0][1]:[n_center,m_center-1],
                 center_numbers.values[1][0]:[n_center-1,m_center],
                 center_numbers.values[1][1]:[n_center,m_center]}

# if the m_center is odd and n_center is even then return 1 x 2 matrix

elif n_center % 2 == 0 and m_center % 2 !=0:
    center_numbers = df.iloc[n_center-1:n_center+1, m_center-1]
    random_center = np.array(random.sample(range(10), (center_numbers.shape[0]*1)))
    df.iloc[n_center-1:n_center+1, m_center-1] = random_center.reshape(center_numbers.shape[0])
    center_numbers = df.iloc[n_center-1:n_center+1, m_center-1]
    starter_carrot = np.max(center_numbers.values)
#     coordinates={center_numbers.values[0]:[n_center-1,m_center-1],
#                  center_numbers.values[1]:[n_center+1,m_center-1],
#                  center_numbers.values[2]:[n_center-1,m_center+1],
#                  center_numbers.values[3]:[n_center+1,m_center+1]}


# if the m_center is even and n_center is odd then return 2 x 1 matrix

elif n_center % 2 != 0 and m_center % 2 == 0:
    center_numbers = df.iloc[n_center-1, m_center-1:m_center+1]   
    random_center = np.array(random.sample(range(10), (1*center_numbers.shape[1])))
    df.iloc[n_center-1, m_center-1:m_center+1]   = random_center.reshape(center_numbers.shape[1])
    center_numbers = df.iloc[n_center-1, m_center-1:m_center+1]   
    starter_carrot = np.max(center_numbers.values)

# if the m_center and n_center is odd then return 1 x 1 matrix
else:
    center_numbers = df.iloc[n_center-1, m_center-1]
    starter_carrot = np.max(center_numbers.values)

# bunny eats the starter carrot which is the max value from the center

print(df)
print(center_numbers)
print("starter_carrot",starter_carrot)
print("coordinates",coordinates[starter_carrot])
print("starter_carrot",df.iloc[coordinates[starter_carrot][1],coordinates[starter_carrot][0]])
starter_coordinates =coordinates[starter_carrot]
print("coordinates",starter_coordinates[0])


def eat_carrots(starter_coordinates):
    column = starter_coordinates[1]
    row = starter_coordinates[0]
    print(df.iloc[column-1,row] )
    
    adjacent_carrots = {
        (df.iloc[column+1,row]): [column+1,row], #bottom
        (df.iloc[column-1,row]): [column-1,row], #top
        (df.iloc[column,row-1]): [column,row-1], #left
        (df.iloc[column,row+1]): [column,row+1]} #right
    
    most_carrot = np.max([df.iloc[column+1,row],
            df.iloc[column-1,row],
            df.iloc[column,row-1],
            df.iloc[column,row+1]])
    print(most_carrot)
    new_starter_coordinate = adjacent_carrots[most_carrot]
    return new_starter_coordinate, most_carrot

new_coordinates = eat_carrots(starter_coordinates)

```
