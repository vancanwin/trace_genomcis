```python
import random, os, math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#objectives: 
# 1. make a garden matrix where the central numbers don't have any repeats
# 2. find the central cell with the most carrots
# 3. find the adjacent cell with most carrots, save coordinates, and number of carrots collected
# 4. repeat until all adjacent cells are 0s

def garden_matrix(N,M):
    # generating garden of carrots

    df = pd.DataFrame(np.random.randint(0,10,size=(N, M)))

    #finding the center part of the matrix


    n_center = int(math.ceil(float(N)/2))
    m_center = int(math.ceil(float(M)/2))

# if the matrix centers are even then return 2 x 2 matrix and make sure the center_numbers don't have repeats

    if N % 2 == 0 and M % 2 == 0:
        center_numbers = df.iloc[n_center-1:n_center+1, m_center-1:m_center+1]    
        random_center = np.array(random.sample(range(10), (center_numbers.shape[0]*center_numbers.shape[1])))
        df.iloc[n_center-1:n_center+1, m_center-1:m_center+1] = random_center.reshape(center_numbers.shape[0],center_numbers.shape[1])
        center_numbers = df.iloc[n_center-1:n_center+1, m_center-1:m_center+1]
        starter_carrot = np.max(center_numbers.values)

        coordinates={center_numbers.values[0][0]:[n_center-1,m_center-1],
                     center_numbers.values[1][0]:[n_center,m_center-1],
                     center_numbers.values[0][1]:[n_center-1,m_center],
                     center_numbers.values[1][1]:[n_center,m_center]}

    # if the m_center is odd and n_center is even then return 1 x 2 matrix

    elif N % 2 == 0 and M % 2 !=0:
        center_numbers = df.iloc[n_center-1:n_center+1, m_center-1]
        random_center = np.array(random.sample(range(10), (center_numbers.shape[0]*1)))
        df.iloc[n_center-1:n_center+1, m_center-1] = random_center.reshape(center_numbers.shape[0])
        center_numbers = df.iloc[n_center-1:n_center+1, m_center-1]
        starter_carrot = np.max(center_numbers.values)
        coordinates={center_numbers.values[0]:[n_center-1,m_center-1],
                     center_numbers.values[1]:[n_center,m_center-1]}


    # if the m_center is even and n_center is odd then return 2 x 1 matrix

    elif N % 2 != 0 and M % 2 == 0:
        center_numbers = df.iloc[n_center-1, m_center-1:m_center+1]   
        print(center_numbers.shape)
        random_center = np.array(random.sample(range(10), (1*center_numbers.shape[0])))
        df.iloc[n_center-1, m_center-1:m_center+1]  = random_center.reshape(center_numbers.shape[0])
        center_numbers = df.iloc[n_center-1, m_center-1:m_center+1]   
        starter_carrot = np.max(center_numbers.values)
        coordinates={center_numbers.values[0]:[n_center-1,m_center-1],
                     center_numbers.values[1]:[n_center-1,m_center]}

    # if the m_center and n_center is odd then return 1 x 1 matrix
    else:
        center_numbers = df.iloc[n_center-1, m_center-1]
        starter_carrot = np.max(center_numbers.values)
        coordinates={center_numbers.values[0]:[n_center-1,m_center-1]}


    print(df)
    print("starter_carrot", df.iloc[coordinates[starter_carrot][0],coordinates[starter_carrot][1]])
    starter_coordinates = coordinates[starter_carrot]
    df.iloc[starter_coordinates[0],starter_coordinates[1]] = 0
    return "coordinates = ", starter_coordinates, "starter_carrot =", starter_carrot, df


def eat_carrots(starter_coordinates,adjusted_df, origin_coordinates, N, M):
    df = adjusted_df
    row = starter_coordinates[0]
    column = starter_coordinates[1]
    
    # In order of left, right, top, bottom
    adjacent_carrots = [[row, column], [row, column], [row, column], [row, column]]
    most_carrot_array = [-1, -1, -1, -1]
    if column > 0:
        most_carrot_array[0] = df.iloc[row, column - 1]
        adjacent_carrots[0] = [row, column - 1]
    if column < (M-1):
        most_carrot_array[1] = df.iloc[row, column + 1]
        adjacent_carrots[1] = [row, column + 1]
    if row > 0:
        most_carrot_array[2] = df.iloc[row - 1, column]
        adjacent_carrots[2] = [row - 1, column]
    if row < (N - 1):
        most_carrot_array[3] = df.iloc[row + 1, column]
        adjacent_carrots[3] = [row + 1, column]
    
    most_carrot = max(most_carrot_array)
    new_starter_coordinate = adjacent_carrots[most_carrot_array.index(most_carrot)]
    if most_carrot == 0:
        print("Shh! Bunny is sleeping!")
        return -1, [row, column], -1, 0, df
    
    
    print("most_carrot = ",most_carrot)
#     print("new_carrot_eat_fx",df.iloc[new_starter_coordinate[0],new_starter_coordinate[1]])
    df.iloc[new_starter_coordinate[0],new_starter_coordinate[1]] = 0
#     print('After the zeros = ',df)
# 
    return "new coordinates: ",new_starter_coordinate, 'most carrots:', most_carrot, df

# gerate the matrix, run the eat_carrot function in a loop and receive all carrots in the list
N,M = 6,5


# print("df before loop",starter_coordinates_carrots[4])

def run_rabbit(N,M):
    starter_coordinates_carrots = garden_matrix(N,M)
    print(starter_coordinates_carrots[1])
    print(starter_coordinates_carrots[4])
    list_carrots = [starter_coordinates_carrots[3]]    
    new_coordinates = eat_carrots(starter_coordinates_carrots[1],starter_coordinates_carrots[4], starter_coordinates_carrots[1], N,M)
    n = 0
    while new_coordinates[3] > 0:
        print(new_coordinates[1])
        print(new_coordinates[4])
        n+=1
        print('round = ',n)

        list_carrots.append(new_coordinates[3])
        new_coordinates = eat_carrots(new_coordinates[1], new_coordinates[4], starter_coordinates_carrots[1], N,M)


    print(sum(list_carrots), list_carrots)
    
run_rabbit(6,5)
```
